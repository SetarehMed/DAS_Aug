### Discription
This script is a small package of augmeting tools which are used in the paper published in JOM, in 2020. For more information about the function of each of the tools refer to the paper. 
After augmeting the images existing in each directory they are saved in a new directory with a code name addressing the type of the applied augmetation.  The augmented images can be searched with the same code name in the directory.

    "Damage Analysis in Dual-Phase Steel Using Deep Learning: Transfer from Uniaxial to Biaxial Straining Conditions by Image Data Augmentation"

     Setareh Medghalchi, Carl F. Kusche, Ehsan Karimi, Ulrich Kerzel & Sandra Korte-Kerze
https://link.springer.com/article/10.1007/s11837-020-04404-0


Setareh Medghalchi - July 2020

```python
# Creat a new Directory for augmented images 
newpath = r'YourDir/Augmented/' 
if not os.path.exists(newpath):
    os.makedirs(newpath)
```

```python
# Read files 
# Activate or deactivate the tools w.r.t your need. 
# Some of the tools 
i = 0
for file in glob.glob('OrgImgDir/*.png'):


    if file.endswith ('.png'):
        image = cv2.imread(file,0)
        
        ##Naming
        name = os.path.splitext(str(file))[0]
        base=os.path.basename(file)
        Org_name = os.path.splitext(base)[0]
        print (Org_name)
        #naming = ntpath.basename(str(file))
        
        
        ## GuassBlr
        aug = iaa.Sometimes(0.5, iaa.GaussianBlur(sigma=2.0)) # Apply gaussian blur to about 50% of all images:
        image_aug = aug.augment_image(image)
        image_aug_seq_Name = ("%s" % (Org_name)) +str(i)+'xcbxb_GuassBlr12.png'
        cv2.imwrite('YourDir/AugmentedImages/'+image_aug_seq_Name ,image_aug)
        
        
        
         ## Flip RL
        aug = iaa.Fliplr(1)
        image_aug = aug.augment_image(image)
        image_aug_seq_Name = ("%s" % (Org_name)) +str(i)+'xcbxb_Flprl11.png'
        cv2.imwrite('YourDir/Augmented/'+image_aug_seq_Name ,image_aug)

        
        ## Rotation
        aug = iaa.Affine(rotate=(0, 180))
        image_aug = aug.augment_image(image)
        image_aug_seq_Name = ("%s" % (Org_name)) +str(i)+'xcbxb_r15.png'
        cv2.imwrite('YourDir/Augmented/'+image_aug_seq_Name ,image_aug)

        
        
        ## Dropout
        aug = aug = iaa.Dropout(p=(0, 0.3)) 
        image_aug = aug.augment_image(image)
        image_aug_seq_Name = ("%s" % (Org_name)) +str(i)+'xcbxb_DrpO11.png'
        cv2.imwrite('YourDir/AugmentedImages/'+image_aug_seq_Name ,image_aug)
    
```

```python
# Read files 
# Activate or deactivate the tools w.r.t your need. 
# All of the tools 
i = 0
for file in glob.glob('OrgImgDir/*.png'):
    
    if file.endswith ('.png'):
        image = cv2.imread(file,0)
        
        ##Naming
        name = os.path.splitext(str(file))[0] # prints the whple directory + naeming
        base=os.path.basename(file) # name + format 
        Org_name = os.path.splitext(base)[0] # only name 
        print (Org_name)
        #naming = ntpath.basename(str(file))
        
        
        ## Rotation
        aug = iaa.Affine(rotate=(0, 180))
        image_aug = aug.augment_image(image)
        image_aug_seq_Name = ("%s" % (Org_name)) +str(i)+'xcbxb_r12.png'
        cv2.imwrite('YourDir/Augmented/'+image_aug_seq_Name ,image_aug)

                
        ## Rotation
        aug = iaa.Affine(rotate=(0, 180))
        image_aug = aug.augment_image(image)
        image_aug_seq_Name = ("%s" % (Org_name)) +str(i)+'xcbxb_r13.png'
        cv2.imwrite('YourDir/Augmented/'+image_aug_seq_Name ,image_aug)

        
        aug = iaa.Affine(rotate=(0, 180))
        image_aug = aug.augment_image(image)
        image_aug_seq_Name = ("%s" % (Org_name)) +str(i)+'xcbxb_r12.png'
        cv2.imwrite('YourDir/AugmentedImages/'+image_aug_seq_Name ,image_aug)

        
        ## Shear
        aug = iaa.Affine(shear=(-16, 16)) # Play arounf with values 
        image_aug = aug.augment_image(image)
        image_aug_seq_Name = ("%s" % (Org_name)) +str(i)+'xcbxb_shr12.png'
        cv2.imwrite('YourDir/Augmented/'+image_aug_seq_Name ,image_aug)

        ## Dropout
        aug = aug = iaa.Dropout(p=(0, 0.3)) 
        image_aug = aug.augment_image(image)
        image_aug_seq_Name = ("%s" % (Org_name)) +str(i)+'xcbxb_DrpO11.png'
        cv2.imwrite('YourDir/Augmented/'+image_aug_seq_Name ,image_aug)
        
        ## Coarse Dropout
        aug  = iaa.CoarseDropout(0.05, size_percent=0.5)  # Drop 5% of all pixels by converting them to black pixels, but do that on a lower-resolution version of the image that has 50% of the original size, leading to 2x2 squares being dropped:
        image_aug = aug.augment_image(image)
        image_aug_seq_Name = ("%s" % (Org_name)) +str(i)+'xcbxb_CrsDrpO13.png'
        cv2.imwrite(' YourDir/Augmented/'+image_aug_seq_Name ,image_aug)

        ##Translation
        aug = iaa.Affine(translate_percent={"x": (-0.2, 0.1), "y": (-0.1, 0.2)}) #Translate images by -20 to 20 pixels on x- and y-axis independently:
        image_aug = aug.augment_image(image)
        image_aug_seq_Name = ("%s" % (Org_name)) +str(i)+'ExSitu_Trns13.png'
        cv2.imwrite('YourDir/Augmented/'+image_aug_seq_Name ,image_aug)


        ## Random pixel generation by translation
        aug = iaa.Affine(translate_percent={"x": -0.20}, mode=ia.ALL, cval=(0, 255)) #When applying affine transformations, new pixels are often generated. then a random brightness between 0 and 255 is used:
        image_aug = aug.augment_image(image)
        image_aug_seq_Name = ("%s" % (Org_name)) +str(i)+'ExSitu_RnndTrns1.png'
        cv2.imwrite('YourDir/Augmented/'+image_aug_seq_Name ,image_aug)


        ## GuassBlr
        aug = iaa.Sometimes(0.5, iaa.GaussianBlur(sigma=4.0)) # Apply gaussian blur to about 50% of all images:
        image_aug = aug.augment_image(image)
        image_aug_seq_Name = ("%s" % (Org_name)) +str(i)+'xcbxb_GuassBlr13.png'
        cv2.imwrite('YourDir/Augmented/'+image_aug_seq_Name ,image_aug)

        
        ## Flip RL
        aug = iaa.Fliplr(1)
        image_aug = aug.augment_image(image)
        image_aug_seq_Name = ("%s" % (Org_name)) +str(i)+'xcbxb_Flprl13.png'
        cv2.imwrite('YourDir/Augmented/'+image_aug_seq_Name ,image_aug)

        
        ## Sharpen
        aug = iaa.Sharpen(alpha=(0.0, 1.0), lightness=(0.75, 2.0))
        image_aug = aug.augment_image(image)
        image_aug_seq_Name = ("%s" % (Org_name)) +str(i)+'xcbxb_hrp11.png'
        cv2.imwrite('YourDir/Augmented/'+image_aug_seq_Name ,image_aug)

        
        ##AdditiveGaussianNoise
        aug = iaa.AdditiveGaussianNoise(scale=(0, 0.05*255))
        image_aug = aug.augment_image(image)
        image_aug_seq_Name = ("%s" % (Org_name)) +str(i)+'xcbxb_AddGuasNoi11.png'
        cv2.imwrite('YourDir/Augmented/'+image_aug_seq_Name ,image_aug)


        ## Random Combination (Sequential)
        #seq1 = iaa.Sequential([
        #    iaa.Flipud(0.1),
        #    iaa.AdditiveGaussianNoise(scale=(50, 80)),
        #    ], random_order=True) # apply augmenters in random order
        #images_aug = seq1(images=image)
        #image_aug_seq_Name = ("%s" % (Org_name)) + str(i)+'xcbxb_seq13'+'.png'
        #cv2.imwrite('/home/setareh/Data/First_network/CORRECT/AugmentedImages/'+image_aug_seq_Name,image_aug)
        

```
